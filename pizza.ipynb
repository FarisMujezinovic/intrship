This notebook loads in data from Pizza places in USA and computers next metrics:
-At what time do people order the most pizzas?
-Average price
-How much revenue is generated per type of pizza on a monthly basis

```python
import pandas as pd
from IPython.display import display
order_details=pd.read_csv("/Users/mujjof/Desktop/ostalo/archive-3/order_details.csv")
orders=pd.read_csv("/Users/mujjof/Desktop/ostalo/archive-3/orders.csv")

merged_order=pd.merge(
    left=order_details[["order_id","quantity","pizza_id"]],
    right=orders[["order_id","date","time"]],
    on=["order_id"],
    how="inner"
)
display(merged_order)
```

```python
merged_order["time"] = pd.to_datetime(merged_order["time"],infer_datetime_format=False).dt.hour
```

```python
#At what time of day are the most pizzas ordered?
merged_order.groupby("time")["order_id"].count()
```

```python

#Analysis 2: Join the pizza_type and pizzas datasets, and from the combined dataset,


pizza_types=pd.read_csv("/Users/mujjof/Desktop/ostalo/archive-3/pizza_types.csv",encoding_errors="ignore")
pizzas=pd.read_csv("/Users/mujjof/Desktop/ostalo/archive-3/pizzas.csv",on_bad_lines="skip")
```

```python
merged_pizza=pd.merge(
    left=pizza_types[["pizza_type_id","name"]],
    right=pizzas[["pizza_id","pizza_type_id","size","price"]],
    on=["pizza_type_id"],
    how="inner"
)
display(merged_pizza)
```

```python
merged_pizza.groupby(["name"])["price"].mean().sort_values(ascending=False)
```

```python
merged_datasets=pd.merge(
    left=merged_order[["order_id","quantity","date","time","pizza_id"]],
    right=merged_pizza[["pizza_id","name","price","size"]],
    on=["pizza_id"],
    how="inner"
)
display(merged_datasets)
```

```python
merged_datasets.groupby(by=["name","size"],as_index=False).agg(
    price=("price", sum)
)
```

```python
merged_datasets.groupby("size")["quantity"].count()
```
