This notebook loads in data from F1 teams and drivers and computers follwing metric:
-total points earned for each team,
-first and last season in F1,
-average number of points for each team,
-first places for constructors,
-second places for constructors,
-third places for constructors,
-most points for drivers,
-number of titles
-average number of points for each driver


```python
import pandas as pd
from IPython.display import display

dfcons=pd.read_csv("/Users/mujjof/Desktop/ostalo/data/untitled folder/constructors_championship_1958-2020.csv")
dfcons.head()
```

```python
#total points earned
dfcons.groupby(["Team"])["Points"].sum().sort_values(ascending=False)
display(dfcons)
```

```python
#first and last season in F1
dfcons.groupby(by="Team",as_index=False).agg(
    first_season=("Year", min),
    last_season=("Year", max),
)
```

```python
#average number of points

dfcons.groupby(["Team"])["Points"].mean().sort_values(ascending=False)



```

```python
#first places for constructors

dfcons[dfcons["Position"] == "1"].groupby("Team").agg(
    wins=("Position", lambda x: x.count())
    ).sort_values(by="wins",ascending=False)
```

```python
#second places for constructors
dfcons[dfcons["Position"] == "2"].groupby("Team").agg(
    wins=("Position", lambda x: x.count())
    ).sort_values(by="wins",ascending=False)
```

```python
#third places for constructors
dfcons[dfcons["Position"] == "3"].groupby("Team").agg(
    wins=("Position", lambda x: x.count())
    ).sort_values(by="wins",ascending=False)
```

```python
import pandas as pd
from IPython.display import display

dfdriver=pd.read_csv("/Users/mujjof/Desktop/ostalo/data/untitled folder/drivers_championship_1950-2020.csv")
dfdriver.head()
```

```python
#most points
dfdriver.groupby(["Name"])["Points"].sum().sort_values(ascending=False)



```

```python
#number of titles

dfdriver[dfdriver["Position"] == "1"].groupby("Name").agg(
    wins=("Position", lambda x: x.count())
    ).sort_values(by="wins",ascending=False)





```

```python
#average number of points
dfdriver.groupby(["Name","Year"])["Points"].mean().sort_values(ascending=False)
```
