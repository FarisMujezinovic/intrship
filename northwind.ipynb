This notebook loads in data and showing us  revenue from the government and  total sales revenue for all category 

```python
import pandas as pd
from IPython.display import display
categories=pd.read_csv("/Users/mujjof/Desktop/ostalo/data/untitled folder/categories.csv",on_bad_lines="skip")
orders=pd.read_csv("/Users/mujjof/Desktop/ostalo/data/untitled folder/orders.csv",on_bad_lines="skip")
products=pd.read_csv("/Users/mujjof/Desktop/ostalo/data/untitled folder/products.csv",on_bad_lines="skip")
suppliers=pd.read_csv("/Users/mujjof/Desktop/ostalo/data/untitled folder/suppliers.csv",on_bad_lines="skip")
products.head()
```

```python
revenue_from_the_government=pd.merge(
    left=products[["supplierID","unitPrice","unitsOnOrder"]],
    right=suppliers[["country","supplierID"]],
    on=["supplierID"],
    how="left"
)
display(revenue_from_the_government)
```

```python

revenue_from_the_government["revenue_by_country"] = revenue_from_the_government["unitPrice"] * revenue_from_the_government["unitsOnOrder"]
prihod_po_drzavama = revenue_from_the_government.groupby("country")["revenue_by_country"].sum().sort_values(ascending=False).reset_index()

display(revenue_from_the_government)

```

```python
#For each category, find the total sales revenue for that category.
revenue_for_that_category=pd.merge(
    left=products[["categoryID","unitPrice","unitsOnOrder"]],
    right=categories[["categoryName","categoryID"]],
    on=["categoryID"],
    how="left"
)
display(revenue_for_that_category)
```

```python

revenue_for_that_category["revenue_by_category"] = revenue_for_that_category["unitPrice"] * revenue_for_that_category["unitsOnOrder"]
prihod_po_kategorijama = revenue_for_that_category.groupby("categoryName")["revenue_by_category"].sum().reset_index()
display(revenue_for_that_category)
```
